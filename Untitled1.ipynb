```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import inv
from numpy.linalg import pinv
from ISLP import load_data
import statsmodels.api as sm
from ISLP.models import summarize, ModelSpec as MS, poly
from sklearn import linear_model as skl
from sklearn import model_selection as skm
from statsmodels.regression.linear_model import OLS
import scipy as sc
```

```python
b0sbs=np.empty(100)
b1sbs=np.empty(100)
b2sbs=np.empty(100)
b0sOLS=np.empty(100)
b1sOLS=np.empty(100)
b2sOLS=np.empty(100)

n=512
tcrit=sc.stats.t.ppf(1 - 0.05/2, n-1)

#OLS
for i in range(100):
    x1=np.random.normal(0,1,n)
    x2=np.random.normal(5,2,n)
    x1=np.sort(x1)
    x2=np.sort(x2)
    b0=2
    b1=3
    b2=4
    w=np.random.normal(0,1,n)
    y=b0+b1*x1+b2*x2+w
    X=np.empty((n,3))
    X[:,0]=np.ones(n)
    X[:,1]=x1
    X[:,2]=x2
    model=sm.OLS(y,X)
    results=model.fit()
    b0sOLS[i],b1sOLS[i],b2sOLS[i]=results.params

#Bootstrap
x1=np.random.normal(0,1,n)
x2=np.random.normal(5,2,n)
x1=np.sort(x1)
x2=np.sort(x2)
b0=2
b1=3
b2=4
w=np.random.normal(0,1,n)
y=b0+b1*x1+b2*x2+w


X=np.empty((n,3))
X[:,0]=np.ones(n)

for i in range(100):   
    index=np.random.choice(np.arange(n),n)   
    X[:,1]=x1[index]
    X[:,2]=x2[index]
    yy=y[index]
    model=sm.OLS(yy,X)
    results=model.fit()
    b0sbs[i],b1sbs[i],b2sbs[i]=results.params


fig,ax=plt.subplots(3,figsize=(10, 8))
fig.suptitle(r"Ordinary least squares confidence intervals for coefficients $b_i (+histograms)$"+"\n"+ r"and bootstrap coefficients $b_i$"+f" for sample size n={n} for y={b0}+{b1}"+r"$x_1$+"f"{b2}"+r"$x_2$",color="RED")
fig.tight_layout()


ax[0].hist(b0sOLS)
ax[0].axvline(x=b0,ymin=0,ymax=5,color='black',label=r"real $_0$")
ax[0].axvline(x=np.mean(b0sOLS),ymin=0,ymax=5,color='red',label=r"$\bar b_0$: OLS mean")
ax[0].axvline(x=np.mean(b0sOLS)-tcrit*np.std(b0sOLS),ymin=0,ymax=5,color='green',label=r"$\bar{b_0}\pm$"+f"{tcrit:.3f}"+r"$\cdot _{b_0}$ (95% Cl)")
ax[0].axvline(x=np.mean(b0sOLS)+tcrit*np.std(b0sOLS),ymin=0,ymax=5,color='green')
ax[0].axvline(x=np.mean(b0sbs),ymin=0,ymax=5,color='yellow',label=r"$b_0$ bootstrap mean")
ax[0].legend()

ax[1].hist(b1sOLS)
ax[1].axvline(x=b1,ymin=0,ymax=5,color='black',label=r"real $_1$")
ax[1].axvline(x=np.mean(b1sOLS),ymin=0,ymax=5,color='red',label=r"$\bar b_1$: OLS mean")
ax[1].axvline(x=np.mean(b1sOLS)-tcrit*np.std(b1sOLS),ymin=0,ymax=5,color='green',label=r"$\bar{b_1}\pm$"+f"{tcrit:.3f}"+r"$\cdot _{b_1}$ (95% Cl)")
ax[1].axvline(x=np.mean(b1sOLS)+tcrit*np.std(b1sOLS),ymin=0,ymax=5,color='green')
ax[1].axvline(x=np.mean(b1sbs),ymin=0,ymax=5,color='yellow',label=r"$b_1$ bootstrap mean")
ax[1].legend()

ax[2].hist(b2sOLS)
ax[2].axvline(x=b2,ymin=0,ymax=5,color='black',label=r"real $_2$")
ax[2].axvline(x=np.mean(b2sOLS),ymin=0,ymax=5,color='red',label=r"$\bar b_2$: OLS mean")
ax[2].axvline(x=np.mean(b2sOLS)-tcrit*np.std(b2sOLS),ymin=0,ymax=5,color='green',label=r"$\bar{b_2}\pm$"+f"{tcrit:.3f}"+r"$\cdot _{b_2}$ (95% Cl)")
ax[2].axvline(x=np.mean(b2sOLS)+tcrit*np.std(b2sOLS),ymin=0,ymax=5,color='green')
ax[2].axvline(x=np.mean(b2sbs),ymin=0,ymax=5,color='yellow',label=r"$b_2$ bootstrap mean")
ax[2].legend()
```
